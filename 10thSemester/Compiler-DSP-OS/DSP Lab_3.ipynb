```python
def sort(x):
    for i in range(len(x)):
        for j in range(len(x)-1):
            if x[j]>x[j+1]:
                temp = x[j]
                x[j] = x[j+1]
                x[j+1] = temp
        return x
x = [34,0,-2,99,10,23,1]
y = sort(x)
print(y)
```

```python
#generate a sinusoid..

import numpy as np
import matplotlib.pyplot as plt
F = 20
Fs = 200
L= 100
x = np.zeros(L)
for n in range(L):
   x[n] = np.sin(2*np.pi*F*n/Fs)
#print(x)
plt.plot(x)
```

```python
#unit impulse

n = range(-10,10)
d = np.zeros(len(n))
d[len(n)//2] = 1
plt.stem(n,d)
```

```python
#unit step

n = range(-10,10)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i]=1
plt.stem(n,d)
```

```python
#exponential function
ur = np.zeros(50)
a = -0.95
for n in range(50):
    ur[n] = a**n
plt.stem(ur)
```

```python
#Ramp function

r = np.zeros(40)
for n in range(40):
    r[n] = n
plt.stem(r)
```
