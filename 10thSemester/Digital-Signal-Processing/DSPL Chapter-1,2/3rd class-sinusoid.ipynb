```python
def sort(x):
    for i in range(len(x)):
        for j in range(len(x) - 1):
            if x[j]>x[j+1]:
                temp = x[j]
                x[j] = x[j+1]
                x[j + 1] = temp
    return x
    
x = [34, 0, -2, 99, 10, 23, 1]
y= sort(x)
print(y)
```

```python
#1. Generate sinusoid
import numpy as np
import matplotlib.pyplot as plt

F = 10 #Set the frequency 
Fs = 400 # Set the sampling frequency
L = 50 # Set the size of line
x = np.zeros(L) # initialize the zero(0)

for n in range(L):
    x[n] = np.sin(2*np.pi*F*n/Fs)

print(x)
plt.plot(x)
plt.stem(x)
```

```python
#unit step
n = range(-10, 11)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i] = 1
plt.stem(n,d)
```

Certainly! Here's a short and easy explanation of the code:

1. `n = range(-10, 11)`: This line creates a range of numbers from -10 to 10 (inclusive), which will serve as the x-axis values for our plot.

2. `d = np.zeros(len(n))`: This line creates an array `d` filled with zeros, having the same length as the range of `n`. The `np.zeros` function is used from the NumPy library.

3. `for i in range(len(n)//2):`: This line starts a loop that iterates over the indices from 0 to half the length of `n`. The `len(n)//2` gives the integer division of the length of `n` by 2, ensuring that the loop only runs for half the length of `n`.

4. `d[len(n)//2+i] = 1 + i`: This line assigns values to `d` starting from the middle index (`len(n)//2`) and moving towards the end of the array (`len(n)//2 + i`). The values assigned are 1 plus the value of `i`, which increases as the loop progresses. This creates a ramp-like shape where the values increment gradually.

5. `plt.stem(n, d)`: This line plots the ramp signal using the stem plot function from the matplotlib library. The `stem` function is used to create a stem plot, where each sample value in `d` is represented as a vertical line above the corresponding value in `n`. This plot visually represents the discrete nature of the ramp signal, where the values gradually increase from the middle index towards the end of the array.

In summary, the code generates a discrete ramp signal with a length of 21 and plots it using the stem plot function. The ramp starts from 1 at the middle index and increases by 1 for each subsequent index, resulting in a ramp-like shape in the plot.

```python
#ramp
n = range(-10, 11)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i] = 1 + i
plt.stem(n,d)
```
